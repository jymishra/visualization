# Distplots in sns

```python
sns.distplot( iris["sepal_length"] , color="skyblue", label="Sepal Length")
sns.distplot( iris["sepal_width"] , color="red", label="Sepal Width")
plt.legend()
#plt.show()
```

```python
sns.distplot( iris["petal_length"] , color="skyblue", label="Petal Length")
sns.distplot( iris["petal_width"] , color="red", label="Petal Width")
plt.legend()
```

# Distplots in Grid

```python
f, axes = plt.subplots(2, 2, figsize=(7, 7), sharex=True)
sns.distplot( iris["sepal_length"] , color="skyblue", ax=axes[0, 0])
sns.distplot( iris["sepal_width"] , color="olive", ax=axes[0, 1])
sns.distplot( iris["petal_length"] , color="gold", ax=axes[1, 0])
sns.distplot( iris["petal_width"] , color="teal", ax=axes[1, 1])
```

# Pairplot with hue as target variable

```python
plt.close()
sns.set_style("whitegrid")
sns.pairplot(iris,hue='species',size=3)
plt.show()
```

# Pairplot wiith Regression Lines

```python
plt.close()
sns.set_style("whitegrid")
sns.pairplot(iris,hue='species',kind='reg',size=3)
plt.show()
```

# Pairplot with ScatterPlot

```python
plt.close()
sns.set_style("whitegrid")
sns.pairplot(iris,hue='species',kind='scatter',size=3)
plt.show()
```

# Pairplot with Histogram

```python
plt.close()
sns.set_style("whitegrid")
sns.pairplot(iris,hue='species',diag_kind='hist',size=3)
plt.show()
```

# diag_kind is always 'auto' by default

```python
plt.close()
sns.set_style("whitegrid")
sns.pairplot(iris,hue='species',diag_kind='auto',size=3)
plt.show()
```
